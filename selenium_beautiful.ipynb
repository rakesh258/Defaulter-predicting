```python
offerings = driver.find_element(By.XPATH, '//article[@class="container-fluid p0"]')
print(len([offerings]))


for offering in [offerings]:
    # Find the offering title
    # offering_title = offering.find_element(By.CLASS_NAME, 'offering-title').text.strip()
    hovers = offering.find_elements(By.XPATH, '//div[@class="row-fluid offering-hover"]')
    #print(len(hovers))
    for j in hovers:
        
        offering_title = j.find_element(By.XPATH, '//h2[@class="offering-title"]')
        print(offering_title.text)
        # Find the location title
        # location_title = offering.find_elements(By.CLASS_NAME, 'mb0')
        location_title = j.find_elements(By.XPATH, '//h3[@class="mb0 mt0 fontweight400"]')

        # Find the location text
        #location_text = offering.find_elements(By.CLASS_NAME, 'location-text')
        location_text = j.find_elements(By.XPATH, '//div[@class="location-text"]')

        for location_title, location_text in zip(location_title, location_text):
            #print(f, b)
            # Print the extracted information
            #print(offering_title)
            print(location_title.text)
            print(location_text.text)
            print()
# Quit the WebDriver
#driver.quit()
```

```python
# Find all offerings
offerings = soup.find('article', class_='container-fluid p0')

for i in [offerings]:
    #print(type(i))
    #print(i)
    offers = i.find_all("div", "row-fluid offering-hover")
    #print(type(offers))
    #print(offers)
    for j in offers:
        # # Find the offering title
        offering_title = j.find('h2', class_="offering-title")
        #print(offering_title.text)
        
        # Find the location title
        location_title = j.find_all('h3', class_='mb0 mt0 fontweight400')
        # Find the location text
        location_text = j.find_all('div', class_='location-text')

        for f, b in zip(location_title, location_text):
            #print(f, b)
            # Print the extracted information
            print(offering_title.text)
            print(f.text)
            print(b.text)
            print()
```

```python
import requests

import re

def get_file_extension(filename):
    # Regular expression pattern to match file extension
    pattern = r'\.([^.]+)$'

    # Search for the file extension using the pattern
    match = re.search(pattern, filename)

    # Extract and return the file extension if found
    if match:
        return match.group(1)
    else:
        return None
    
    is_pdf = get_file_extension(link.get('href'))
    if is_pdf == "pdf":
        print(link.get('href'))
```
